```python
#Data Cleaning 
data2['City'] = data2['City'].str.replace('Gurgaon','Delhi-NCR')
data2['City'] = data2['City'].str.replace('Noida','Delhi-NCR')
data2['City'] = data2['City'].str.replace('Faridabad','Delhi-NCR')
data2['City'] = data2['City'].str.replace('New Delhi','Delhi-NCR')
data2['City'] = data2['City'].str.replace('Ghaziabad','Delhi-NCR')
```

```python
#Setting the country code to India and making new dataframe accordingly
Indian_country_code=1
new=data2[data2['Country Code']==Indian_country_code]
```

```python
## new dataframe will have restaurants only in india
new.head(3)
```

```python
dfd=new[new.City=='Delhi-NCR']
dfr=new[new.City!='Delhi-NCR']
```

```python
#dfd.Cuisines.nunique()
#dfr.Cuisines.nunique()
def Dictionary(array):
    dictionary={}
    for i in array:
        if ',' not in i:
            if i.strip() in dictionary:
                dictionary[i.strip()]=dictionary.get(i.strip())+1
            else:
                dictionary[i.strip()]=1   
        else:
            string=i.strip().split(',')
            for j in string:
                if j.strip() in dictionary:
                    dictionary[j.strip()]=dictionary.get(j.strip())+1
                else:
                    dictionary[j.strip()]=1
    return dictionary 
```

```python
#dic has Cuisines in Delhi NCR with their count
dic=Dictionary(dfd['Cuisines'])
dfr.Cuisines.fillna('Not Known',inplace=True)
#dic2 has cuisines in all of India with their count
dic2=Dictionary(dfr['Cuisines'])
```
